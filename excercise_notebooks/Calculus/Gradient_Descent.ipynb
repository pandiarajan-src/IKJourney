# Introduction to Gradient Descent

Gradient descent is an optimization algorithm used to minimize the cost function of a model. It is commonly used in machine learning to update the model parameters iteratively until the optimal solution is found. The idea behind gradient descent is to calculate the gradient of the cost function with respect to the parameters and update them in the opposite direction of the gradient until the minimum is reached.

The gradient descent algorithm can be described as follows:

1. Initialize the model parameters with random values.
2. Calculate the cost function using the current parameters.
3. Calculate the gradient of the cost function with respect to each parameter.
4. Update each parameter by subtracting the product of the learning rate and the gradient from the current parameter.
5. Repeat steps 2-4 until the cost function converges to a minimum.

The learning rate is a hyperparameter that controls the step size at each iteration. A high learning rate can result in convergence to a minimum quickly, but it may overshoot and oscillate around the minimum. On the other hand, a low learning rate can result in a slow convergence to the minimum.

There are different variants of gradient descent, including batch gradient descent, minibatch gradient descent, and stochastic gradient descent. These variants differ in the amount of data used to compute the gradient at each iteration and how the model parameters are updated.

# Gradient Descent with Fixed Learning Rate

Let's start by implementing gradient descent with a fixed learning rate in Python.


We will use the following cost function:

$
f(x) = (x - 5)^2 + 20
$


Let us the plot the function first!

```python
# x
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return (x - 5) ** 2 + 20

# Create an array of x values
x = np.linspace(-10, 20, 1000)

# Evaluate f(x) for each value of x
y = f(x)

# Plot f(x) versus x
plt.plot(x, y)

# Add labels and a title to the plot
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) versus x')

# Display the plot
plt.show()
```

It appears to be a simple quadratic function with a minimum at x = 5. We will use gradient descent to find the minimum of this function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the cost function
def cost_function(x):
    return (x - 5) ** 2 + 20

# Define the gradient of the cost function
def gradient(x):
    return 2 * (x - 5)

# Initialize the model parameter with a random value
x = np.random.randn()
# x = -5

# Set the learning rate
learning_rate = 0.1

# Set the maximum number of iterations
max_iterations = 100

# Initialize the list to store the cost values
cost_values = []

# Iterate until convergence or max iterations
for i in range(max_iterations):
    # Calculate the cost and gradient at the current parameter
    cost = cost_function(x)
    grad = gradient(x)

    # Print the progress
    if i % 10 == 0:
        print("Iteration {}: x = {}, cost = {}".format(i, x, cost))

    # Update the parameter
    x = x - learning_rate * grad

    # Store the cost value
    cost_values.append(cost)



# Plot the cost values
plt.plot(cost_values)
plt.xlabel("Iterations")
plt.ylabel("Cost")
plt.show()
```

In the above code, we first define the cost function and its gradient. We then initialize the model parameter with a random value and set the learning rate and maximum number of iterations. We iterate until max iterations and calculate the cost and gradient at each iteration. We update the parameter by subtracting the product of the learning rate and gradient from the current parameter. We also store the cost value at each iteration and print the progress. Finally, we plot the cost values to visualize the convergence.

Let's observe how x changes after each iteration of Gradient Descent!

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x)
def f(x):
    return (x - 5) ** 2 + 20

# Define the gradient of f(x)
def grad_f(x):
    return 2 * (x - 5)

# Define the gradient descent algorithm
def gradient_descent(f, grad_f, x_init, alpha, num_iterations):
    x = x_init
    x_list = [x]
    for i in range(num_iterations):
        x = x - alpha * grad_f(x)
        x_list.append(x)
    return x_list

# Set the hyperparameters
alpha = 0.1
num_iterations = 5

# Run gradient descent on f(x)
x_init = 15
x_list = gradient_descent(f, grad_f, x_init, alpha, num_iterations)

# Create an array of x values
x = np.linspace(-2, 15, 100)

# Evaluate f(x) for each value of x
y = f(x)

# Plot f(x) versus x
plt.plot(x, y)

# Plot the gradient descent iterations
for i in range(len(x_list) - 1):
    x1 = x_list[i]
    y1 = f(x1)
    x2 = x_list[i + 1]
    y2 = f(x2)
    plt.plot([x1, x2], [y1, y2], 'ro-')
    plt.text(x1, y1 + 0.5, round(y1, 2))

# Label the final cost value
x_final = x_list[-1]
y_final = f(x_final)
plt.text(x_final, y_final + 0.5, round(y_final, 2))

# Add labels and a title to the plot
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient descent iterations on f(x)')

# Display the plot
plt.show()
```
