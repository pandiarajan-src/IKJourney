```python
# You have a DataFrame df with columns 'A', 'B', and 'C', 
# and you want to replace all occurrences of the value 0 in column 'A' with the mean of non-zero values in that column. 
# Which of the following code snippets is correct?
import numpy as np
import pandas as pd


# Create a DataFrame with 100 rows and columns 'A', 'B', and 'C'
df = pd.DataFrame(np.random.randint(0, 100, size=(100,3)), columns=["A", "B", "C"])
zero_rows = pd.DataFrame(np.zeros((9, 3), dtype=int), columns=["A", "B", "C"])
df = pd.concat([df, zero_rows], axis=0)
print(df)

zero_columns = df[df['A'] == 0]
# print(zero_columns)

# Calculate the mean of non-zero values in column 'A'
mean_non_zero_A = df.loc[df["A"] != "0", 'A'].mean()
print(mean_non_zero_A)

# Replace 0 values in column 'A' with the calculated mean
df.loc[df['A'] == 0, 'A'] = int(mean_non_zero_A)
# df = df['A'].replace(0, df['A'][df['A'] != 0].mean())
print(df)

```

```python
# Merging 2 Dataframe on a common column

import pandas as pd
import numpy as np

df1 = pd.DataFrame(np.random.randint(10, 100, size=(100, 3)), columns=["A","B", "C"])
df2 = pd.DataFrame(np.random.randint(10, 100, size=(100, 2)), columns=["C", "D"])

print(df1)
print(df2)

df = pd.merge(df1, df2, on="C", how="inner")
print(df)
```

```python
# Write a Python program using Pandas to create a DataFrame 
# from the given data and calculate the average rating of movies directed by Christopher Nolan.

import pandas as pd

data = {
    'Title': ['Inception', 'Dunkirk', 'Interstellar', 'The Prestige', 'Memento'],
    'Director': ['Christopher Nolan', 'Christopher Nolan', 'Christopher Nolan', 'Christopher Nolan', 'Christopher Nolan'],
    'Rating': [8.8, 7.9, 8.6, 8.5, 8.4]
}

df = pd.DataFrame(data)
print(f"given input: \n {df}")

# average rating of movies directed by Christopher Nolan
df1 = df.groupby("Director")["Rating"].mean()
print(f"\naverage rating of movies directed by Christopher Nolan: \n{df1}")

df2 = df.groupby("Director").agg({"Rating":"mean"})
print(f"\naverage rating of movies directed by Christopher Nolan: \n{df2}")


```

```python
# Given the following DataFrame, 
# write a Python program using Pandas to extract all rows 
# where the price is greater than or equal to 20,000 
# and sort them in descending order by price.

# data = {
#     'Product': ['Laptop', 'Desktop', 'Tablet', 'Phone', 'Smartwatch'],
#     'Price': [25000, 12000, 8000, 22000, 5000]
# }

import pandas as pd

data = {
    'Product': ['Laptop', 'Desktop', 'Tablet', 'Phone', 'Smartwatch'],
    'Price': [25000, 12000, 8000, 22000, 5000]
}

df = pd.DataFrame(data)
print(f"\n given data: \n {df}")

df1 = df[df["Price"] >= 20000].sort_values("Price", ascending=False)
print(f"\n List of products whose price is greater than 20K in descending order: \n {df1}")
```

```python
# Write a Python program using Pandas to create a DataFrame from the given data 
# and find the total revenue and average price of items sold in each store.

# data = {
#     'Store': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
#     'Item': ['Apple', 'Banana', 'Orange', 'Grape', 'Apple', 'Banana', 'Orange', 'Grape'],
#     'Price': [50, 20, 30, 60, 55, 22, 33, 65],
#     'Quantity': [10, 12, 15, 16, 20, 25, 30, 35]
# }

import pandas as pd

data = {
    'Store': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Item': ['Apple', 'Banana', 'Orange', 'Grape', 'Apple', 'Banana', 'Orange', 'Grape'],
    'Price': [50, 20, 30, 60, 55, 22, 33, 65],
    'Quantity': [10, 12, 15, 16, 20, 25, 30, 35]
}
df1 = pd.DataFrame(data)
print(f"\n given data: \n {df1}")

df1["revenue"] = df1["Price"] * df1["Quantity"]

print(f"\n total revenue and average price of items sold in each store:")
result = df1.groupby("Store").agg({"revenue": "sum", "Price": ["mean", "median"]})
print(result)
```

```python
# Write a Python program using Pandas to create a DataFrame from the given data 
# and calculate the total amount spent by each customer.

# data = {
#     'Customer': ['Alice', 'Bob', 'Alice', 'Alice', 'Bob', 'Bob', 'Alice', 'Bob'],
#     'Item': ['Pen', 'Pencil', 'Notebook', 'Eraser', 'Pen', 'Pencil', 'Notebook', 'Eraser'],
#     'Price': [10, 5, 50, 20, 10, 5, 50, 20],
#     'Quantity': [3, 4, 2, 5, 10, 6, 1, 2]
# }

import pandas as pd

data = {
    'Customer': ['Alice', 'Bob', 'Alice', 'Alice', 'Bob', 'Bob', 'Alice', 'Bob'],
    'Item': ['Pen', 'Pencil', 'Notebook', 'Eraser', 'Pen', 'Pencil', 'Notebook', 'Eraser'],
    'Price': [10, 5, 50, 20, 10, 5, 50, 20],
    'Quantity': [3, 4, 2, 5, 10, 6, 1, 2]
}
df = pd.DataFrame(data)

df["Item_Spend"] = df["Price"] * df["Quantity"]
print(df)
print(df.groupby("Customer").agg({"Item_Spend": "sum"}))
```

```python
# Write a Python program using Pandas to create a DataFrame from the given data 
# and find the total amount spent per item, 
# as well as the total amount spent by all customers.

import pandas as pd

data = {
    'Customer': ['Alice', 'Bob', 'Alice', 'Alice', 'Bob', 'Bob', 'Alice', 'Bob'],
    'Item': ['Pen', 'Pencil', 'Notebook', 'Eraser', 'Pen', 'Pencil', 'Notebook', 'Eraser'],
    'Price': [10, 5, 50, 20, 10, 5, 50, 20],
    'Quantity': [3, 4, 2, 5, 10, 6, 1, 2]
}

df = pd.DataFrame(data)

df["item_spend"] = df['Price'] * df['Quantity']
print(df)

print(f"\n total amount spent per item: \n {df.groupby("Item").agg({"item_spend": "sum"})}")
print(f"\n total amount spent per item: \n {df.groupby("Customer").agg({"item_spend": "sum"})}")

```

```python
#  Write a Python program using Pandas to create a DataFrame from the given data 
# and calculate the monthly average temperature difference between two cities.

import pandas as pd

data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'City_A_Temp': [20, 22, 25, 27, 30, 32, 35, 34, 30, 28, 24, 21],
    'City_B_Temp': [10, 12, 15, 17, 20, 22, 25, 24, 20, 18, 14, 11]
}

df = pd.DataFrame(data)

df['avg_temp_bn_cities'] = (df['City_A_Temp'] + df['City_B_Temp']) / 2
df['temp_diff_bn_cities'] = df['City_A_Temp'] - df['City_B_Temp']
print(df)
```
