```python
import numpy as np
import matplotlib.pyplot as plt

loaded_4_sided_die_probability = {1: 1/5, 2: 2/5, 3: 1/5, 4: 1/5} 
four_sided_dice = np.array([1, 2, 3, 4])

all_outcomes = np.array([(x, y) for x in four_sided_dice for y in four_sided_dice])

sums = np.zeros(len(all_outcomes))
probabilities = np.zeros(len(all_outcomes))

for i in range(all_outcomes.shape[0]):
    x, y = all_outcomes[i][0], all_outcomes[i][1]
    sums[i] = x + y
    probabilities[i] = loaded_4_sided_die_probability[x] * loaded_4_sided_die_probability[y]

# Calculate the unique sums and their probabilities
unique_sums, unique_counts = np.unique(sums, return_counts=True)

plt.bar(unique_sums, unique_counts, color='skyblue', edgecolor='black')
plt.title('Probability Mass Function of the Sum of Two Loaded 4-Sided Dice')
plt.xlabel('Sum')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the loaded probabilities
# P(1) = 1/5, P(2) = 2/5, P(3) = 1/5, P(4) = 1/5
loaded_probabilities = {
    1: 1/5,
    2: 2/5,
    3: 1/5,
    4: 1/5
}

# Generate all possible outcomes for two throws
outcomes = [1, 2, 3, 4]
all_outcomes = [(x, y) for x in outcomes for y in outcomes]

# Calculate the sum of the two throws and their probabilities
sums = []
probabilities = []

for x, y in all_outcomes:
    sums.append(x + y)
    probability = loaded_probabilities[x] * loaded_probabilities[y]
    probabilities.append(probability)

# Calculate the unique sums and their probabilities
unique_sums = sorted(set(sums))
pmf = [sum(probabilities[i] for i in range(len(sums)) if sums[i] == s) for s in unique_sums]

pmf1 = []
for s in unique_sums:
    for i in range(len(sums)):
        if sums[i] == s:
            pmf1.append(probabilities[i])

# Plot the PMF as a histogram
plt.bar(unique_sums, pmf, color='skyblue', edgecolor='black')
plt.title('Probability Mass Function of the Sum of Two Loaded 4-Sided Dice')
plt.xlabel('Sum')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Return the PMF for inspection
unique_sums, pmf, pmf1
```

```python
# Excercise 2

# Define the number of sides and possible outcomes of a fair 6-sided die
sides = 6
outcomes = np.arange(1, sides + 1)

# Generate all possible outcomes for two throws
all_outcomes = [(x, y) for x in outcomes for y in outcomes]

# Calculate the sum of the two throws
sums = [x + y for x, y in all_outcomes]

# Calculate the probability of each sum (since the die is fair, each outcome is equally likely)
# Total possible outcomes for two throws = 6 * 6 = 36
total_outcomes = sides ** 2
probability_each_sum = 1 / total_outcomes

# Unique sums and their probabilities
unique_sums, counts = np.unique(sums, return_counts=True)
pmf = counts * probability_each_sum

# Plot the PMF as a histogram
plt.bar(unique_sums, pmf, color='lightblue', edgecolor='black')
plt.title('Probability Mass Function of the Sum of Two Fair 6-Sided Dice')
plt.xlabel('Sum')
plt.ylabel('Probability')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Return the PMF for inspection
unique_sums, pmf
```

```python
import numpy as np
import matplotlib.pyplot as plt

def loaded_dice(p):
    """Simulate a loaded 6-sided dice"""
    return np.random.choice(p)

def throw_dice(p, num_throws):
    """Throw the loaded dice twice and calculate the sum"""
    sums = []
    for _ in range(num_throws):
        throw1 = loaded_dice(p)
        throw2 = loaded_dice(p)
        sums.append(throw1 + throw2 + 2)  # Add 2 because indices start at 0
    return sums

def calculate_mean_and_variance(sums):
    """Calculate the mean and variance of the sums"""
    mean = np.mean(sums)
    variance = np.var(sums)
    return mean, variance

# Define the loading of the dice
loadings = [
    [1, 2, 3, 4, 5, 6],  # Fair dice
    [1, 1, 2, 3, 4, 5, 6],  # Loaded on 1
    [1, 2, 2, 3, 4, 5, 6],  # Loaded on 2
    [1, 2, 3, 3, 4, 5, 6],  # Loaded on 3
    [1, 2, 3, 4, 4, 5, 6],  # Loaded on 4
    [1, 2, 3, 4, 5, 5, 6],  # Loaded on 5
    [1, 2, 3, 4, 5, 6, 6],  # Loaded on 6    
]

# Simulate throwing the dice
num_throws = 100000
means = []
variances = []
for loading in loadings:
    sums = throw_dice(loading, num_throws)
    mean, variance = calculate_mean_and_variance(sums)
    means.append(mean)
    variances.append(variance)

# Print the results
for i, (mean, variance) in enumerate(zip(means, variances)):
    print(f"Loading {i+1}: Mean = {mean:.2f}, Variance = {variance:.2f}")

# Plot the results
plt.plot(means, label="Mean")
plt.plot(variances, label="Variance")
plt.xlabel("Loading")
plt.ylabel("Value")
plt.title("Mean and Variance of Sums")
plt.legend()
plt.show()
```
