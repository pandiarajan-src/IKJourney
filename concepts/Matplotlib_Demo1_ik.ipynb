```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
type(iris)
```

```python
iris.describe("")
```

```python
# Scatter plot of sepal length and sepal width
plt.scatter(iris["sepal_length"], iris["sepal_width"])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Scatter plot of sepal length (vs) sepal width")
plt.show()
```

```python
# Box plot

plt.boxplot(x='petal_length', data=iris)
plt.title("Boxplot of petal length")
plt.show()
```

```python
# histogram of petal length

plt.hist(iris["petal_length"], bins=10)
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.title(("Histogram of petal length"))
plt.show()
```

```python
# How to create a pie chart
species_count  = iris["species"].value_counts()
plt.pie(species_count, labels=species_count.index)
plt.title("Pie chart distribution of species")
plt.show()
```
